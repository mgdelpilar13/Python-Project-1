```python
# @title Year vs Total Tested

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Year']
  ys = series['Total Tested']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = Manhattan_schools.sort_values('Year', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('School Level')):
  _plot_series(series, series_name, i)
  fig.legend(title='School Level', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Year')
_ = plt.ylabel('Total Tested')
```

```python
# @title School Level

from matplotlib import pyplot as plt
import seaborn as sns
Manhattan_schools.groupby('School Level').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```
